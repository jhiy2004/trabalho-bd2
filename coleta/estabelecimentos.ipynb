```python
import pandas as pd
from pathlib import Path
import re
```

```python
datasets = Path(r"C:\Users\jhiy2\OneDrive\Documentos\trabalho-bd2\dados")
```

```python
anos = [i for i in range(2016, 2023)]

maiores_cidades = [
    "JOINVILLE",
    "FLORIANOPOLIS",
    "BLUMENAU",
    "SAO JOSE",
    "ITAJAI",
    "CHAPECO",
    "PALHOCA",
    "CRICIUMA",
    "JARAGUA DO SUL",
    "LAGES",
    "BRUSQUE",
    "BALNEARIO CAMBORIU",
    "TUBARAO",
    "CAMBORIU",
    "NAVEGANTES",
    "RIO DO SUL",
    "ITAPEMA",
    "ARMAZEM",
    "CONCORDIA",
    "CACADOR",
    "ITUPORANGA",
    "SAO MIGUEL DO OESTE",
    "RIO NEGRINHO",
    "SAO BONIFACIO",
    "BIGUACU",
    "SAO FRANCISCO DO SUL",
    "ICARA",
    "GASPAR",
    "MAFRA",
    "IRATI",
    "PALMITOS",
    "LAJEADO GRANDE",
    "PIRATUBA",
    "IBIRAMA",
    "ENTRE RIOS",
    "ORLEANS",
    "BALNEARIO PICARRAS",
    "PORTO UNIAO",
    "BRACO DO NORTE",
    "SAO LOURENCO DO OESTE",
    "PORTO BELO",
    "PONTE ALTA DO NORTE",
]


dfs = [pd.read_csv(datasets / f"estab_{i}.csv", sep=";") for i in anos]
```

```python
dfs[0].columns
```

```python
for i, df in enumerate(dfs):
    df["ano"] = anos[i]
    df["Município"] = df["Município"].apply(lambda s : re.sub(r'^\d+\s*', '', s))
    df.drop(df.index[-1], inplace=True)
    df.drop(columns="Total", inplace=True)
    df = df.query("Município in @maiores_cidades")

    dfs[i] = df.reset_index(drop=True)
```

```python
dfs[0].head()
```

```python
cidades_df = set(dfs[0]["Município"].unique())
cidades_faltando = set(maiores_cidades) - cidades_df

print(cidades_faltando)
```

```python
df_final = pd.concat(dfs)
```

```python
df_pivot = df_final.pivot_table(
    index="Município",
    columns="ano",
    values=["POSTO DE SAUDE", "CENTRO DE SAUDE/UNIDADE BASICA", "PRONTO SOCORRO GERAL", "HOSPITAL GERAL"],
    aggfunc="sum"   # ou "first" se não houver duplicatas
)
```
